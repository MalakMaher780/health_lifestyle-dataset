```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

health_data = pd.read_csv('health_lifestyle_dataset.csv')
health_data.head()
```

```python
# Check missing values before cleaning
print("Missing values before cleaning:")
print(health_data.isnull().sum())

# Fill missing numeric values with mean
health_data.fillna(health_data.mean(numeric_only=True), inplace=True)

# Fill missing categorical values with mode
health_data.fillna(health_data.mode().iloc[0], inplace=True)

# Check again after cleaning
print("\nMissing values after cleaning:")
print(health_data.isnull().sum())

```

```python
# Select one column
ages = health_data['age']

# Select multiple columns
subset = health_data[['age', 'daily_steps', 'sleep_hours']]

# Select rows 0–4
first_rows = health_data.iloc[0:5]
```

```python
# Example: calculate total weekly exercise hours
health_data['WeeklyExercise'] = health_data['daily_steps'] * 7

# Example: simple health score (custom feature)
health_data['HealthScore'] = (health_data['sleep_hours'] * 10) - health_data['bmi']
health_data.head()
```

```python
# Select people with high disease risk
high_risk = health_data[health_data['disease_risk'] == 'High']

# Select people older than 40
above_40 = health_data[health_data['age'] > 40]
subset = health_data[['age', 'bmi', 'sleep_hours']]
high_risk.head()
above_40.head()

```

```python
# Sort by Age
health_data.sort_values('age', ascending=True, inplace=True)

# Group by disease risk and get averages
grouped = health_data.groupby('disease_risk')[['age', 'bmi', 'sleep_hours']].mean()
print(grouped)
```

```python
# Rename a column
health_data.rename(columns={'ExerciseHours': 'DailyExercise'}, inplace=True)

# Drop unwanted columns
health_data.drop(['Unnamed: 0'], axis=1, inplace=True, errors='ignore')

# Reorder columns
health_data = health_data[['age', 'bmi', 'daily_steps', 'sleep_hours', 'disease_risk']]
health_data.head()
health_data.columns

```

```python
# Add a new calculated column
health_data['BMI_Category'] = pd.cut(
    health_data['bmi'],
    bins=[0, 18.5, 24.9, 29.9, 100],
    labels=['Underweight', 'Normal', 'Overweight', 'Obese']
)
health_data.head()
```

```python
plt.figure(figsize=(8,6))
plt.hist(health_data['bmi'], bins=15, color='mediumseagreen', edgecolor='black', alpha=0.7)

plt.title('Distribution of BMI in the Dataset')
plt.xlabel('BMI')
plt.ylabel('Number of People')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
bmi_by_gender = health_data.groupby('gender')['bmi'].mean()

bmi_by_gender.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Average bmi by gender')
plt.xlabel('gender')
plt.ylabel('Average bmi')
plt.show()
```

```python
health_data['BMI_Category'].value_counts().plot(
    kind='bar', color='lightgreen', edgecolor='black'
)
plt.title('Count of People by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(health_data['bmi'], health_data['sleep_hours'], alpha=0.5, color='purple', edgecolors='black', s=60)
plt.title('BMI vs. Sleep Hours', fontsize=14)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Sleep Hours', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()

```

```python
smoker_counts = health_data['smoker'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral','skyblue'])
plt.title('Proportion of Smokers vs Non-Smokers')
plt.show()
```

```python
# Scatter plot: age vs Systolic Blood Pressure
plt.figure(figsize=(8,6))
plt.scatter(health_data['age'], health_data['systolic_bp'], color='mediumseagreen', alpha=0.7)

# Add labels and title
plt.title('Relationship between AGE and Systolic Blood Pressure')
plt.xlabel('AGE')
plt.ylabel('Systolic Blood Pressure (mmHg)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(health_data['age'], health_data['diastolic_bp'], color='cornflowerblue', alpha=0.7)

plt.title('Relationship between AGE and Diastolic Blood Pressure')
plt.xlabel('AGE')
plt.ylabel('Diastolic Blood Pressure (mmHg)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(health_data['age'], health_data['systolic_bp'], color='red', alpha=0.6, label='Systolic BP')
plt.scatter(health_data['age'], health_data['diastolic_bp'], color='blue', alpha=0.6, label='Diastolic BP')

plt.title('AGE vs Blood Pressure (Systolic & Diastolic)')
plt.xlabel('age')
plt.ylabel('Blood Pressure (mmHg)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(health_data['cholesterol'], bins=15, color='cornflowerblue', edgecolor='black', alpha=0.7)

plt.title('Distribution of Cholesterol in the Dataset')
plt.xlabel('Cholesterol')
plt.ylabel('Number of People')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
health_data['disease_risk'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'gold', 'lightblue']
)
plt.title('Disease Risk Distribution')
plt.ylabel('')  # hide y-label
plt.show()

```

```python
health_data.boxplot(column='sleep_hours', by='disease_risk', grid=False)
plt.title('Sleep Hours by Disease Risk')
plt.suptitle('')  # removes extra title
plt.xlabel('Disease Risk')
plt.ylabel('Sleep Hours')
plt.show()
```
